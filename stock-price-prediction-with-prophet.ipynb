```python
import yfinance as yf
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import dexplot as dxp

import plotly.express as px
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)
from plotly import tools
import plotly.figure_factory as ff



from prophet.plot import plot_plotly, plot_components_plotly
import datetime as dt
from prophet import Prophet

import os
import glob

```

```python
yf_tsla = yf.Ticker("TSLA")
tsla = yf_tsla.history(period='max')
tsla['Date'] = tsla.index
df = tsla[tsla.index > '2020-05-01'][['Date', 'Close']]
df.rename(columns={'Close': 'y', 'Date': 'ds'}, inplace=True)
```

```python
tsla_m = Prophet()
tsla_m.fit(df)
future_prices = tsla_m.make_future_dataframe(periods=60,freq='D')
forecast = tsla_m.predict(future_prices)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
## Forecasting for Next Month Dec 2017 & Jan 2018
### y represents the truth values and yhat forecasted values. 
fig = ([
    go.Scatter(x=df['ds'], y=df['y'], name='y'),
    go.Scatter(x=forecast['ds'], y=forecast['yhat'], name='yhat'),
    go.Scatter(x=forecast['ds'], y=forecast['yhat_upper'], fill='tonexty', mode='none', name='upper'),
    go.Scatter(x=forecast['ds'], y=forecast['yhat_lower'], fill='tonexty', mode='none', name='lower')
])


py.iplot(fig)
```
